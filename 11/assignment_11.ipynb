## compute distance
- feature : $n \times m$, center : $1 \times m$, distance : $n \times 1$
- $n$ : number of data, $m$ : number of features

```python
def compute_distance(feature, center):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    # center한개에 모든 feature의 distance를 계산해야함
    # feature은 n*m matrix고 center은 1*m matrix
    temp_dist = feature - center
    distance = np.sqrt(np.sum(temp_dist**2,axis=0))
    print( distance)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return distance
```

## compute centroid
- feature : $n \times m$, label_feature : $n \times 1$, value_label : $1 \times 1$, centroid : $1 \times m$
- $n$ : number of data, $m$ : number of features

```python
def compute_centroid(feature, label_feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    
    centroid=np.zeros(2)
    for i in range(feature.shape[0]):
        if(label_feature.all()==label):
            #x좌표
            centroid[0] += feature[i][0]
            #y좌표
            centroid[1]+=feature[i][1]
        centroid = centroid/feature.shape[0]
   


    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return centroid
```
