## compute distance
- feature : $n \times m$, center : $1 \times m$, distance : $n \times 1$
- $n$ : number of data, $m$ : number of features

```python
def compute_distance(feature, center):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    # center한개에 모든 feature의 distance를 계산해야함
    # feature은 n*m matrix고 center은 1*m matrix
    temp_dist = (feature - center)**2
    distance = np.power(temp_dist[:,0]+temp_dist[:,1], 0.5)
    # print( distance)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return distance
```

## compute centroid
- feature : $n \times m$, label_feature : $n \times 1$, value_label : $1 \times 1$, centroid : $1 \times m$
- $n$ : number of data, $m$ : number of features

```python
def compute_centroid(feature, label_feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    
    # centroid=np.zeros(2)
    # print(centroid.shape)
    # for i in range(feature.shape[0]):
    #     if(label_feature.all()==label):
    #         #x좌표
    #         centroid[0] += feature[i][0]
    #         #y좌표
    #         centroid[1]+=feature[i][1]
    #     centroid = centroid/feature.shape[0]
    target=feature[np.where(label==label_feature)]
    centroid=target.sum(axis=0)/len(target)
    

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return centroid
```

## compute label
- distance : $n \times k$, label_feature : $n \times 1$
- $n$ : number of data, $k$ : number of clusters

```python
def compute_label(distance):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    #distance 중에 값이 최소가 되는 인덱스
    
    label_feature = np.argmin(distance)
    
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    
    return label_feature
```

## the number of clusters $K = 2$

```python
number_cluster          = 2
number_iteration        = 2     # you can modify this value
loss_iteration_02       = np.zeros(number_iteration)
centroid_iteration_02   = np.zeros((number_iteration, number_cluster, number_feature))
label_feature_02        = np.random.randint(0, number_cluster, size=(number_data))
```

```python
# ++++++++++++++++++++++++++++++++++++++++++++++++++
# complete the blanks
#
distance            = np.zeros(shape=(number_data, number_cluster))
centroid            = np.zeros(shape=(number_cluster, number_feature))
for i in range(number_iteration):
    loss_sum =0
    for j in range(number_cluster):
        centroid[j] = compute_centroid(feature, label_feature_02, j)
        distance[:,j] = compute_distance(feature,centroid[j])

    centroid_iteration_02[i]=centroid
    
    # distance[i] = compute_distance(feature,centroid)
   
    label_feature_02 = compute_label(distance)
    loss_iteration_02[i] = sum(distance[range(feature.shape[0]),label_feature_02])/feature.shape[0]

    # for k in range(number_data):
        # label = label_feature_02[k]
        # point = distance[label][k]
        # loss_sum += point
        
    # loss_iteration_02[i] = loss_sum/feature.shape[0]
    
    # print(loss_iteration_02)

#
# ++++++++++++++++++++++++++++++++++++++++++++++++++
```
